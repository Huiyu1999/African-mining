```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Load the GeoJSON file
geojson_path = '/Users/kd6801/Desktop/Mining-Project/Congo-Data/Congo-villages.geojson'
villages = gpd.read_file(geojson_path)

# Create a Point for the specific coordinates
specific_point = Point(26.8314956853973, -1.9091487367551845)

# Plot the data
fig, ax = plt.subplots(figsize=(10, 10))
villages.plot(ax=ax, color='blue', markersize=1, label='Villages')

# Plot the specific point
plt.plot(specific_point.x, specific_point.y, marker='o', color='red', markersize=10, label='Specific Point')

# Customize the plot
ax.set_title('Village Distribution')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Add a legend
plt.legend()

# Show the plot
plt.show()


```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Path to the shapefile
shapefile_path = '/Users/kd6801/Desktop/Mining-Project/Congo-Data/Congo-protected/protected_areas.shp'

# Load the shapefile
protected_areas = gpd.read_file(shapefile_path)

# Create a Point for the specific coordinates
specific_point = Point(26.8314956853973, -1.9091487367551845)

# Plot the data
fig, ax = plt.subplots(figsize=(10, 10))
protected_areas.plot(ax=ax, color='green', edgecolor='black', alpha=0.5, label='Protected Areas')

# Plot the specific point
plt.plot(specific_point.x, specific_point.y, marker='o', color='red', markersize=10, label='Specific Point')

# Customize the plot
ax.set_title('Protected Areas and Specific Point')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Path to the shapefile
waterways_shapefile_path = '/Users/kd6801/Desktop/Mining-Project/Congo-Data/Congo-water/hotosm_cod_waterways_lines_shp.shp'

# Load the shapefile
waterways = gpd.read_file(waterways_shapefile_path)

# Create a Point for the specific coordinates
specific_point = Point(26.8314956853973, -1.9091487367551845)

# Plot the data
fig, ax = plt.subplots(figsize=(10, 10))
waterways.plot(ax=ax, color='blue', linewidth=0.5, label='Waterways')

# Plot the specific point
plt.plot(specific_point.x, specific_point.y, marker='o', color='red', markersize=10, label='Specific Point')

# Customize the plot
ax.set_title('Waterways and Specific Point')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
